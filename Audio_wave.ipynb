```python
import wave
import matplotlib.pyplot as plt
import numpy as np
```

```python
s = wave.open("Audio.wav",'rb')
s
```

```python
print("No.of.channels:",s.getnchannels())
print("Sample Width:",s.getsampwidth())
print("Frame Rate:",s.getframerate())
print("No.of.Frames:",s.getnframes())
print("Parameters:",s.getparams())
```

```python
t_audio = s.getnframes() / s.getframerate()
t_audio
```

```python
f = s.readframes(-1)
print(type(f))
print(len(f) / 2)
```

```python
s1 = wave.open("Audio.wav",'wb')
s1
```

```python
s1.setnchannels(1)
s1.setsampwidth(2)
s1.setframerate(44100)
```

```python
s1.writeframes(f)
s1.close()
```

```python
signal = np.frombuffer(f,dtype=np.int16)
signal
```

```python
times = np.linspace(0, t_audio, num=len(signal))
times
```

```python
plt.figure(figsize=(15,5))
plt.plot(times,signal)
plt.title("Audio Signal")
plt.ylabel("Signal Wave")
plt.xlabel("Time (sec)")
plt.xlim(0,t_audio)
plt.show()
```
